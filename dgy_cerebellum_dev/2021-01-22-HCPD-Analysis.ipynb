# Cerebellum

```python
maps = HCPD.getAllMaps(HCPD.getCerebellumMap)
np.save(os.path.join(DATA_DIR, 'Maps-HCPD-Cerebellum'), arr = maps)
```

```python
maps = np.load(os.path.join(DATA_DIR, 'Maps-HCPD-Cerebellum.npy'), allow_pickle = True).item()
```

```python
# HCPD.setOutliersAsNan
# New strategy to deal with outliers
# Rather than eliminating all values of a particular postion where outliers appear,
# other values are kept with outliers being nan.
for age in HCPD.AGES:
    for i, sub in enumerate(maps[age].copy()):
        indices = HCPD.getRowOutliersByIQR(sub, n = 2)
        maps[age][i, indices] = np.nan
```

```python
mean_maps = {age: np.nanmean(maps[age], axis = 0) for age in maps}
array = np.row_stack(mean_maps.values())
normals = HCPD.getNormalIndices(array, HCPD.getRowOutliersByNan)
array = array[:, normals]
```

```python
matrix, _ = HCPD.generateMatrix(array, False)
HCPD.showMatrix(matrix, vmin = 0.7)
```
