```python
import os
import pickle as pkl
from os.path import join as pjoin
from scipy.cluster.hierarchy import dendrogram
from matplotlib import pyplot as plt
from cxy_visual_dev.lib.predefine import proj_dir

anal_dir = pjoin(proj_dir, 'analysis')
work_dir = pjoin(anal_dir, 'plot/dendrogram')
if not os.path.isdir(work_dir):
    os.makedirs(work_dir)
```

```python
fname = 'HCPY-M+T_MMP-vis2-LR_zscore1-split_PCA-subj_hac-C1'
out_file = pjoin(work_dir, f'{fname}.jpg')
data = pkl.load(open(pjoin(anal_dir, f'cluster/{fname}.pkl'), 'rb'))
plt.figure(figsize=(12, 4.8))
dendrogram(data['Z'], labels=data['label'], leaf_rotation=45)
plt.savefig(out_file)
```
